```
from keras.preprocessing.text import one_hot

docs= ['glass of orange Juice',
       'bottle of mango juice',
       'glass of mango shake',
       'drink bottle of banana shake',
       'I want a glass of cold water',
       'The king and the queen', 
       'man and woman']
```

```
vocab_size=10000
encoded_docs= [one_hot(d, vocab_size) for d in docs]
print(encoded_docs)
```

## Word Embedding

```
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding
import numpy as np

embedding_len= 5
max_doc_len=10 # total columns in the embedding matrix we want 
encoded_docs= pad_sequences(encoded_docs, truncating='post', padding='post', maxlen=max_doc_len) # zeros are padded "post" in the embedding matrix
encoded_docs
```

```
model= Sequential()
model.add(Embedding(vocab_size, embedding_len, input_length=max_doc_len))
model.compile('rmsprop', 'mse')
output= model.predict(encoded_docs)
output
```

```
output[0] # embedding for first line - 'glass of orange juice'
#the last 6 rows are same because the zero padded in the embedding sequence.(10-4==6 || embedding col - no. of words in the line of one-hot = zeros padded)
```

```
output[0][0] # embedding for word 'glass'
```
